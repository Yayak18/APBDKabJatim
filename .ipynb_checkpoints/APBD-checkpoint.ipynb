### Perbandingan APBD Kabupaten/Kota Se-Jawa Timur

Untuk memahami gambaran kinerja fiskal masing-masing daerah di Jawa Timur.  APBD merupakan instrumen penting bagi Pemerintah Daerah dalam rangka mewujudkan pelayanan dan peningkatan kesejahteraan masyarakatnya. <br>
Analisa sederhana ini menggunakan data dengan periode waktu 4 tahun, yakni tahun 2021 hingga 2024 dengan menggunakan seluruh populasi kabupaten/kota di Jawa timur, yakni 38 kabupaten/kota. <br>
Pemetaan dilakukan terbatas pada beberapa aspek antara lain: <br>
1 Komposisi APBD Kabupaten/Kota; <br>
2 Perhitungan Rasio PAD terhadap PDRB; <br>
3 Pertumbuhan APBD; <br>
4 Pemetaan kapasitas fiskal tiap daerah di Jawa Timur; <br>
5 Pemetaan Belanja Pegawai (konteks: jelang penerapan ps 147 (1) UU 1/2022 HKPD). <br>
Analisa sederhana ini dilakukan menggunakan bahasa pemograman Python. <br>

Sumber data APBD diperoleh dari [Portal SIKD Kementrian Keuangan](https://djpk.kemenkeu.go.id/portal/data/apbd?periode=9&tahun=2023&provinsi=13&pemda=38)<br>
Sumber data PDRB diperoleh dari [Portal BPS](https://djpk.kemenkeu.go.id/portal/data/apbd?periode=9&tahun=2023&provinsi=13&pemda=38)

Langkah awal sebelum dilakukan pemetaan adalah install packages python kemudian membuat daftar nama Kabupaten/Kota dalam bentuk array.

```python
import math
import numpy as np
import pandas as pd

nama_daerah = [
    "Bangkalan", "Banyuwangi", "Batu", "Blitar", "Bojonegoro", "Bondowoso",
    "Gresik", "Jember", "Jombang", "Kediri", "Kota_Blitar", "Kota_Kediri",
    "Kota_Madiun", "Kota_Malang", "Kota_Mojokerto", "Kota_Pasuruan", 
    "Kota_Probolinggo", "Lamongan", "Lumajang", "Madiun", "Magetan", "Malang", 
    "Mojokerto", "Nganjuk", "Ngawi", "Pacitan", "Pamekasan", "Pasuruan", 
    "Ponorogo", "Probolinggo", "Sampang", "Sidoarjo", "Situbondo", "Sumenep", 
    "Surabaya", "Trenggalek", "Tuban", "Tulungagung"
]
for i in nama_daerah:
    print(i)
```

Data APBD yang didapat dari [Portal SIKD Kemenkeu](https://djpk.kemenkeu.go.id/portal/data/apbd?periode=9&tahun=2023&provinsi=13&pemda=38) merupakan himpunan data berformat `.CSV`dengan pola penamaan `kabupaten/kota.csv`. Kumpulan file ini kemudian digabung menjadi 1 dataset untuk memudahkan analisa. 

```python
data = pd.DataFrame()
for i in nama_daerah:
    folder = f'./data/apbd_2023/{i}.csv'
    df = pd.read_csv(folder)
    df = df.drop(columns=["realisasi", "persentase"])
    df.rename(columns={"akun": "Daerah"}, inplace=True)
    df.rename(columns={"anggaran": i}, inplace=True)
    df = df.T  
    if len(data) != 0:
        df = df.drop(["Daerah"])
    data = pd.concat([data, df]) 
    print(data)
```
